```python
import numpy as np
import imutils
import dlib
import cv2
import pafy
import os
```

```python
def video1(message):
    
    cap = cv2.VideoCapture()
    test = cap.open(message)

    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

    counter = 0
    while True:
        buffer, frame = cap.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        faces = detector(gray)
        for face in faces:
            x1 = face.left()
            y1 = face.top()
            x2 = face.right()
            y2 = face.bottom()
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 3)

            landmarks = predictor(gray, face)

            for n in range(0, 68):
                x = landmarks.part(n).x
                y = landmarks.part(n).y
                cv2.circle(frame, (x, y), 4, (255, 0, 0), -1)
        
            #extraction
            cv2.imwrite("Database_images/User" + '.'+ str(counter) + ".jpg",gray[y1:y2,x1:x2])
            counter = counter + 1
            cv2.imshow("Frame", frame)
         
        key = cv2.waitKey(1)
        if key == 27:
            break
            
    cap.release()
    cv2.destroyAllWindows()
    return test



print()
print("If Video from Youtube press 1 \nIf video from any PC press 2 \nIf you want to terminate press 3 ")

i=int(str(input("enter your choice:")))
    
if i==1:
    print("enter url:")
    url = input()
    video = pafy.new(url)
    video_quality1 = video.getbest(preftype="mp4")
    capture = cv2.VideoCapture()
    capture.open(video_quality1.url)
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
    counter = 0
    while True:
        buffer, frame = capture.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = detector(gray)
        for face in faces:
            x1 = face.left()
            y1 = face.top()
            x2 = face.right()
            y2 = face.bottom()
            a = cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 3)
            
            landmarks = predictor(gray, face)
            
            for n in range(0, 68):
                x = landmarks.part(n).x
                y = landmarks.part(n).y
                cv2.circle(frame, (x, y), 4, (255, 0, 0), -1)
            
            #extracting faces
            cv2.imwrite("Database_images/User" + '.'+ str(counter) + ".jpg",gray[y1:y2,x1:x2])
            counter += 1
            cv2.imshow("Frame", frame)
        key = cv2.waitKey(100)
        if key == 27:
            break
    capture.release()
    cv2.destroyAllWindows()
    
    
elif i==2:
    file = input("enter path with file name and extension:")
    video1(file)
    
else:
    print("Terminated.")
      
```
